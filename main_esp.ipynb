## 6. Escribe una funci√≥n que devuelva el factorial de un n√∫mero.

```python
def factorial(n):
    x = 1
    for i in range(1, n+1):
        x *= i
    return x
```

```python
# F√≥rmula factorial
# n! = n * (n - 1) *...*1

# Este c√≥digo define una funci√≥n llamada "factorial" que toma una entrada "n". La funci√≥n utiliza un bucle for para iterar a trav√©s del rango de n√∫meros
# desde 1 hasta n+1. Para cada n√∫mero en ese rango, multiplica el valor actual de x por el n√∫mero en el rango. Al final del bucle,
# la funci√≥n devuelve el valor final de x, que ser√° el factorial del n√∫mero de entrada "n".

# El factorial de un entero positivo n es el producto de todos los enteros positivos menores o iguales a n.
# Por ejemplo, el factorial de 6 (escrito "6!") es 6 * 5 * 4 * 3 * 2 * 1 = 720.

# As√≠ que esta funci√≥n toma una entrada de cualquier entero positivo y devuelve el factorial de ese n√∫mero.
```

## 8. Escribe una funci√≥n que devuelva la moda de una lista, es decir: el elemento que aparece m√°s veces.

`NOTE: No se debe usar count... üßê`

```python
def mode_counter(arr):
    
    counter = {}
    for i in arr:
        if i in counter:
            counter[i] += 1
        else:
            counter[i] = 1
    
    max_count = 0
    max_value = arr[0]
    
    for value in counter:
        count = counter[value]
        if count > max_count:
            max_count = count
            max_value = value
            
    return max_value
```

```python
# Para testear tu funci√≥n 
test_mode(mode_counter)
```

## 9. Escribe una funci√≥n que calcule la desviaci√≥n est√°ndar de una lista.
`NOTE: no utilices librer√≠as ni ninguna funci√≥n ya construida. üòâ`

```python
def st_dev(list_sd):

    # Media
    sum = 0
    n = 0
    for x in list_sd:
        sum += x
        n += 1
    avg = sum / n
    
    # Valor promedio de (x - avg)¬≤
    sum2 = 0
    for x in list_sd:
        sum2 += (x - avg) ** 2
    var = sum2 / (n - 1)  # Varianza
    
    #return math.sqrt(var)
    return var ** 0.5   # Ra√≠z cuadrada
```

```python
# Para testear tu funci√≥n
test_stdev(st_dev)
```

```python
def pangram(string):
    abc = "abcdefghijklmnopqrstuvwxyz"
    ABC = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    
    for i in range(26):
        if abc[i] not in string and ABC[i] not in string:
            return False
    return True
```

## 11. Escribe una funci√≥n que reciba una cadena de palabras separadas por comas y devuelva una cadena de palabras separadas por comas ordenadas alfab√©ticamente.

`NOTA: Puedes usar sorted pero no split y definitivamente no join! ü§™`

```python
def sort_alpha(string):

    # Obtener lista de palabras
    words = []
    word = ""
    for i in string:
        if i != ',':
            word += i
        else:
            words.append(word)
            word = ""
    words.append(word)
    
    words = sorted(words)
    
    # Convertir a string
    result = ""
    n = 0
    for j in words:
        n += 1
        
    for i in range(n):
        result += words[i]
        if i < n-1:
            result += ","
            
    return result
```

```python
# Para testear tu funci√≥n 
test_alpha(sort_alpha)
```

## 12. Escribe una funci√≥n para verificar si una contrase√±a dada es fuerte (al menos 8 caracteres, al menos una min√∫scula, al menos una may√∫scula, al menos un n√∫mero y al menos un car√°cter especial). Deber√≠a devolver True si es fuerte y False si no lo es.

```python
def check_pass(password):

    # Verificar longitud
    length = 0
    for _ in password:
        length += 1
    if length < 8:
        return False
        
    # Verificar min√∫scula
    tiene_minuscula = False
    for char in password:
        if 'a' <= char <= 'z':
            tiene_minuscula = True
            break
    if not tiene_minuscula:
        return False
        
    # Verificar may√∫scula
    tiene_mayuscula = False
    for char in password:
        if 'A' <= char <= 'Z':
            tiene_mayuscula = True
            break
    if not tiene_mayuscula:
        return False
        
    # Verificar n√∫mero
    tiene_numero = False
    for char in password:
        if '0' <= char <= '9':
            tiene_numero = True
            break
    if not tiene_numero:
        return False
        
    # Verificar car√°cter especial
    tiene_especial = False
    for char in password:
        if not ('a' <= char <= 'z' or 'A' <= char <= 'Z' or '0' <= char <= '9'):
            tiene_especial = True
            break
    if not tiene_especial:
        return False
        
    return True
```

```python
# Para testear tu funci√≥n
test_pass(check_pass)
```
